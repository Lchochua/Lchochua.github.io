## 1 Random Variables and Probability Distributions

Basic concepts of probability theory:

- **Outcomes**: Mutually exclusive results of a random process (only one can occur).
- **Probability**: The long-run proportion an outcome occurs after many trials.
- **Sample space**: All possible outcomes of a random variable.
- **Event**: A subset of the sample space (one or more outcomes).

A **random variable** is a numerical summary of random outcomes. It can be:

- **Discrete**: Takes specific values (e.g., 0, 1).
- **Continuous**: Can take any value in a range.

### 🎲 Dice Roll: Probability & Cumulative Distribution

When rolling a fair six-sided die, each outcome (1–6) has an equal probability of 1/6.

**PDF and CDF of a Dice Roll**

| Outcome                | 1   | 2   | 3   | 4   | 5   | 6   |
|------------------------|-----|-----|-----|-----|-----|-----|
| Probability (PDF)      | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |
| Cumulative Probability | 1/6 | 2/6 | 3/6 | 4/6 | 5/6 | 1   |

The **PDF** shows the chance of each specific result, while the **CDF** shows the probability of getting *that value or less*.

````r
# generate the vector of cumulative probabilities 
cum_probability <- cumsum(probability) 

# Set background to white. 
par(bg = "white")

# plot the probabilites 
plot(cum_probability, 
     xlab = "Outcomes", 
     ylab="Cumulative Probability",
     main = "Cumulative Probability Distribution",
     pch=20) 
````

### 🎯 Binomial Distribution from Bernoulli Trials

It is a well-known result that the number of successes **k** in a Bernoulli experiment follows a **binomial distribution**, denoted as:

$$
k \sim B(n, p)
$$

The probability of observing **k** successes in the experiment \( B(n, p) \) is given by:

$$
f(k) = P(k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k} = \frac{n!}{k!(n - k)!} \cdot p^k \cdot (1 - p)^{n - k}
$$

Here, $\binom{n}{k}$ is the **binomial coefficient**.



In **R**, we can calculate this using the `dbinom()` function:

```r
dbinom(x = 5, size = 10, prob = 0.5)
```

This computes the probability of getting **k = 5** successes in **n = 10** trials, with **p = 0.5**, i.e., \( P(k = 5) \).


````r
dbinom(x = 5,
       size = 10,
       prob = 0.5) 
````

## 📊 Expected Value, Mean and Variance

The **expected value** of a random variable is, loosely, the long-run average value of its outcomes when the number of repeated trials is large.

For a **discrete random variable**, the expected value is a **weighted average** of its possible outcomes, where the weights are the corresponding probabilities.


#### 🔹 Key Concept: Expected Value and the Mean

Suppose a random variable \( Y \) takes on \( k \) possible values \( y_1, y_2, \ldots, y_k \), with corresponding probabilities \( p_1, p_2, \ldots, p_k \).

Then the **expected value** of \( Y \), denoted \( E(Y) \), is defined as:

$$
E(Y) = y_1p_1 + y_2p_2 + \cdots + y_kp_k = \sum_{i=1}^{k} y_i p_i
$$

This is also known as the **mean** of \( Y \), and often written as \( \mu_Y \).

The notation \( \sum_{i=1}^{k} y_i p_i \) means "the sum of \( y_i p_i \) for \( i \) running from 1 to \( k \)".


### 🎲 Expected Value of a Dice Roll

In the dice example, the random variable \( D \) takes on 6 possible values:

$$
d_1 = 1, \; d_2 = 2, \; \ldots, \; d_6 = 6
$$

Assuming a fair die, each of the 6 outcomes has a probability of \( \frac{1}{6} \).  
The expected value \( E(D) \) can be calculated by hand:

$$
E(D) = \frac{1}{6} \sum_{i=1}^{6} d_i = 3.5
$$

So, $E(D)$ is simply the **average of the natural numbers from 1 to 6**, since all weights $p_i$ are $\frac{1}{6}$.

This can also be easily computed in **R** using the `mean()` function:



````r
# compute mean of natural numbers from 1 to 6
mean(1:6)
````

#### 🔄 Pseudo-Random Numbers and `set.seed()` in R


Sequences of random numbers generated by R are **pseudo-random numbers** — they are not *truly* random, but closely approximate the behavior of truly random sequences.

Since this approximation is good enough for most applications, we simply refer to them as **random numbers**.



In general, random numbers are generated by **pseudo-random number generators (PRNGs)**.  
The PRNG in R works by applying a deterministic operation on a value — typically, the value previously generated by the PRNG.

However, when using the PRNG for the first time, there is no prior value. That’s where a **seed** comes in:

- A **seed** is the initial value that starts the sequence.
- Each seed leads to a unique sequence of numbers.
- In R, we use `set.seed()` to set this value.

```r
set.seed(123)
```

✅ **Why is this useful?**

### 📐 Variance and Standard Deviation  


The **variance** of a discrete random variable $Y$, denoted $\sigma_Y^2$, is:

$$
\sigma_Y^2 = \text{Var}(Y) = \mathbb{E}[(Y - \mu_Y)^2] = \sum_{i=1}^{k} (y_i - \mu_Y)^2 p_i
$$

The **standard deviation** of $Y$ is $\sigma_Y$, the square root of the variance.  
Its units are the same as those of $Y$.

---

### 🔍 Sample vs Population Variance in R

The variance defined above is the **population variance**, which is **not directly implemented** in R.  
Instead, R’s `var()` function calculates the **sample variance**:

$$
s_Y^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

This differs from the **population variance**, which is:

$$
\text{Var}(Y) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \mu_Y)^2
$$

The **sample variance** measures how $n$ observations vary around the sample mean $\bar{y}$,  
while **population variance** describes the dispersion of the entire population around the mean $\mu_Y$.

---

### 🎲 Dice Example

For a fair die $D$, where $\mu_D = 3.5$, the population variance is:

$$
\text{Var}(D) = \frac{1}{6} \sum_{i=1}^{6} (d_i - 3.5)^2 = 2.92
$$

This value is **different** from the sample variance $s^2$ computed using `var()` in R.



````r
var(1:6)
````

### 📈 Probabilities, Expected Value, and Variance of a Continuous Random Variable

Let $f_Y(y)$ denote the **probability density function (PDF)** of a continuous random variable $Y$.  
The probability that $Y$ falls between $a$ and $b$ (where $a < b$) is:

$$
P(a \leq Y \leq b) = \int_a^b f_Y(y) \, dy
$$

We also have:

$$
P(-\infty \leq Y \leq \infty) = 1 \quad \text{and therefore} \quad \int_{-\infty}^{\infty} f_Y(y) \, dy = 1
$$


In the continuous case, the **expected value** of $Y$ is the **probability-weighted average** of its values,  
using integrals instead of sums:

$$
E(Y) = \mu_Y = \int_{-\infty}^{\infty} y f_Y(y) \, dy
$$

The **variance** is the expected value of the squared deviation from the mean:

$$
\text{Var}(Y) = \sigma_Y^2 = \int_{-\infty}^{\infty} (y - \mu_Y)^2 f_Y(y) \, dy
$$


### 🔍 Example: Continuous Random Variable with PDF

Consider the continuous random variable $X$ with probability density function (PDF):

$$
f_X(x) = \frac{3}{x^4}, \quad x > 1
$$



#### ✅ Validity of the PDF

We verify it's a valid PDF by checking if the area under the curve equals 1:

$$
\int_1^{\infty} f_X(x) \, dx = \int_1^{\infty} \frac{3}{x^4} \, dx
= \lim_{t \to \infty} \int_1^t \frac{3}{x^4} \, dx
= \lim_{t \to \infty} \left[ -\frac{1}{x^3} \right]_1^t
= -\left( \lim_{t \to \infty} \frac{1}{t^3} - 1 \right)
= 1
$$

So $f_X(x)$ is a valid probability density function.



#### 📊 Expected Value

We compute the expectation:

$$
E(X) = \int_1^{\infty} x \cdot f_X(x) \, dx = \int_1^{\infty} x \cdot \frac{3}{x^4} \, dx
= \int_1^{\infty} \frac{3}{x^3} \, dx
= \left[ -\frac{3}{2x^2} \right]_1^{\infty}
= \frac{3}{2}
$$



#### 🧮 Variance

We use the formula $\text{Var}(X) = E(X^2) - [E(X)]^2$. First, compute $E(X^2)$:

$$
E(X^2) = \int_1^{\infty} x^2 \cdot f_X(x) \, dx = \int_1^{\infty} x^2 \cdot \frac{3}{x^4} \, dx
= \int_1^{\infty} \frac{3}{x^2} \, dx
= \left[ -\frac{3}{x} \right]_1^{\infty}
= 3
$$

Now compute variance:

$$
\text{Var}(X) = E(X^2) - [E(X)]^2 = 3 - \left( \frac{3}{2} \right)^2 = \frac{3}{4}
$$



### 🧠 Summary

- The area under the PDF is 1 ✅  
- The expected value is $E(X) = \frac{3}{2}$  
- The variance is $\text{Var}(X) = \frac{3}{4}$  

While we did all this analytically, this can get tedious or impossible when no closed-form solution exists.



### 💡 Tip: Use R for Integration

In **R**, we can automate this using the `integrate()` function.  
Define $f_X(x)$, $x \cdot f_X(x)$, and $x^2 \cdot f_X(x)$ as R functions and compute integrals directly!


````r
# define functions
f <- function(x) 3 / x^4
g <- function(x) x * f(x)
h <- function(x) x^2 * f(x)
````

````r
# compute area under the density curve
area <- integrate(f, 
                 lower = 1, 
                 upper = Inf)$value
area 
````

````r
# compute E(X)
EX <- integrate(g,
                lower = 1,
                upper = Inf)$value
EX
````

````r
# compute Var(X)
VarX <- integrate(h,
                  lower = 1,
                  upper = Inf)$value - EX^2 
VarX
````

### 📈 The Normal Distribution

The **normal distribution** is one of the most important probability distributions, especially due to its role in the **Central Limit Theorem**.

Normal distributions are **symmetric** and **bell-shaped**, fully defined by their **mean** $\mu$ and **standard deviation** $\sigma$.  
We denote a normal distribution as:

$$
\mathcal{N}(\mu, \sigma^2)
$$

The probability density function (PDF) of a normal distribution is:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
$$



### 🧮 The Standard Normal Distribution

For the **standard normal distribution**, we have:

- $\mu = 0$
- $\sigma = 1$
- Random variable often denoted as $Z$
- Denoted as: $Z \sim \mathcal{N}(0, 1)$

The standard normal **PDF** is commonly written as $\phi(c)$  
The standard normal **CDF** is written as $\Phi(c)$

So:

- $\phi(c) = \Phi'(c)$  
- $\Phi(c) = P(Z \leq c)$

### 📐 Computing Probabilities Involving Normal Random Variables

Suppose $Y$ is normally distributed with **mean** $\mu$ and **variance** $\sigma^2$:

$$
Y \sim \mathcal{N}(\mu, \sigma^2)
$$

To compute probabilities, we **standardize** $Y$ using:

$$
Z = \frac{Y - \mu}{\sigma}
$$



Let $c_1$ and $c_2$ be two values with $c_1 < c_2$, and define:

- $d_1 = \frac{c_1 - \mu}{\sigma}$
- $d_2 = \frac{c_2 - \mu}{\sigma}$

Then we can express the following probabilities using the standard normal CDF $\Phi(\cdot)$:

- $P(Y \leq c_2) = P(Z \leq d_2) = \Phi(d_2)$  
- $P(Y \geq c_1) = P(Z \geq d_1) = 1 - \Phi(d_1)$  
- $P(c_1 \leq Y \leq c_2) = P(d_1 \leq Z \leq d_2) = \Phi(d_2) - \Phi(d_1)$

## 📊 The Chi-Squared Distribution

The **chi-squared distribution** is widely used in econometrics, especially for **hypothesis testing** and **regression diagnostics**.

The sum of $M$ **squared independent standard normal** variables follows a chi-squared distribution with $M$ degrees of freedom:

$$
Z_1^2 + Z_2^2 + \cdots + Z_M^2 = \sum_{m=1}^{M} Z_m^2 \sim \chi^2_M \quad \text{with} \quad Z_m \overset{i.i.d.}{\sim} \mathcal{N}(0,1)
$$



A $\chi^2$-distributed random variable with $M$ degrees of freedom has:

- **Mean**: $M$  
- **Mode**: $M - 2$ (for $M \geq 2$)  
- **Variance**: $2 \cdot M$



#### 🔢 Example

If:

$$
Z_1, Z_2, Z_3 \overset{i.i.d.}{\sim} \mathcal{N}(0,1)
$$

Then:

$$
Z_1^2 + Z_2^2 + Z_3^2 \sim \chi^2_3
$$

## 📘 The Student *t* Distribution

Let $Z$ be a standard normal variate and $W \sim \chi^2_M$ a chi-squared random variable with $M$ degrees of freedom.  
Assume that $Z$ and $W$ are independent. Then the random variable

$$
X := \frac{Z}{\sqrt{W / M}} \sim t_M
$$

follows a **Student *t* distribution** with $M$ degrees of freedom.



### ✨ Properties

- The shape of a $t_M$ distribution depends on $M$.
- $t$ distributions are **symmetric** and **bell-shaped**, resembling a normal distribution.
- As $M$ increases, the $t$ distribution becomes increasingly close to the **standard normal distribution**.
- The $t_\infty$ distribution is the **standard normal** distribution.

This approximation is reasonable even for $M \geq 30$.



### 📈 Moments

A $t_M$ distributed variable $X$ has:

- **Expectation** if $M > 1$:
  $$
  \mathbb{E}(X) = 0
  $$
  
- **Variance** if $M > 2$:
  $$
  \text{Var}(X) = \frac{M}{M - 2}
  $$

The Student *t* distribution is often used in inference when the population variance is unknown.


Let us plot some $t$ distributions with different $M$ and compare them to the standard normal distribution.

## 📊 The F Distribution

The **F distribution** arises from the **ratio of two independent** $\chi^2$-distributed random variables, each divided by their respective degrees of freedom.



Let $W \sim \chi^2_M$ and $V \sim \chi^2_n$, with $W$ and $V$ independent. Then the quantity

$$
\frac{W / M}{V / n} \sim F_{M, n}
$$

follows an **F-distribution** with:

- **Numerator degrees of freedom**: $M$  
- **Denominator degrees of freedom**: $n$

We denote this distribution by $F_{M,n}$.

The **F-distribution** was introduced by George Snedecor and named in honor of **Sir Ronald Fisher**.



### 🔍 Support and Shape

- The support of an $F_{M,n}$ random variable is $\mathbb{R}_{\geq 0}$.
- The distribution is right-skewed, especially for small $n$.



### 🧪 Example: Tail Probability

Suppose $Y \sim F_{3, 14}$ and we are interested in computing:

$$
P(Y \geq 2)
$$

This can be done in R using the `pf()` function, setting `lower.tail = FALSE`:


This gives the **right-tail probability** — the mass in the tail beyond the value 2.



You can also use `qf()` for quantiles and `rf()` to simulate random values from the F-distribution.


````r
pf(2, df1 = 3, df2 = 14, lower.tail = F)
````

We can visualize this probability by drawing a line plot of the related density and adding a color shading with `polygon()`.

````r
# Set background to white. 
par(bg = "white")

# define coordinate vectors for vertices of the polygon
x <- c(2, seq(2, 10, 0.01), 10)
y <- c(0, df(seq(2, 10, 0.01), 3, 14), 0)

# draw density of F_{3, 14}
curve(df(x ,3 ,14), 
      ylim = c(0, 0.8), 
      xlim = c(0, 10), 
      ylab = "Density",
      main = "Density Function")

# draw the polygon
polygon(x, y, col = "orange")
````

### 🔁 Relationship to Other Distributions

The **F-distribution** is related to many other distributions. An important special case arises in **econometrics** when the **denominator degrees of freedom** are large.

In this case, the $F_{M,n}$ distribution can be approximated by $F_{M,\infty}$, which is simply the distribution of a chi-squared random variable divided by its degrees of freedom:

$$
\frac{W}{M} \sim F_{M,\infty}, \quad \text{where } W \sim \chi^2_M
$$

This approximation is useful when $n$ is large, and simplifies the analysis.
