## Implementing the OLS Estimator in R

Ordinary Least Squares (OLS) is one of the fundamental techniques in regression analysis, widely used in econometrics and statistics. In this exercise, we will **implement the OLS estimator** step by step in **R**, using matrix operations.

### The OLS Estimator Formula

The OLS estimator for $\hat{\beta}$ is given by:

$$
\hat{\beta} = (X'X)^{-1} X' y.
$$

This equation involves three matrix operations, which we can efficiently implement in **R**:

- **Transpose**: The expression $X'$ is written as `t(X)` in R.
- **Matrix multiplication**: The expression $X'X$ is translated as `t(X) %*% X`.
- **Inverse**: The matrix inversion $(X'X)^{-1}$ is computed using `solve(t(X) %*% X)`.

### Estimating the Variance of the Error Term

We have not covered this material yet but still let implement the following calculations.

The formula for the estimated variance of the error term is:

$$
\hat{\sigma}^2 = \frac{1}{n - k - 1} \hat{u}' \hat{u}
$$


To store this variable as a scalar instead of a $1 \times 1$ matrix, we use:

```r
sigsqhat <- as.numeric( t(uhat) %*% uhat / (n - k - 1) )
```

#### Standard Error of the Regression

The standard error of the regression (SER) is its square root:

$$
\hat{\sigma} = \sqrt{\hat{\sigma}^2}
$$

#### OLS Variance-Covariance Matrix

The estimated OLS variance-covariance matrix is:

$$
\widehat{\text{Var}}(\hat{\beta}) = \hat{\sigma}^2 (X'X)^{-1}
$$


#### Standard Errors of Parameter Estimates

Finally, the standard errors of the parameter estimates are the square roots of the main diagonal of $\text{Var}(\beta)$.
