## Introduction

Classical statistics is a fundamental approach in econometrics and economic analysis. It is based on the **analogy principle**, which states that if an econometrician is interested in some characteristic of a population, they should use the analogy in the sample. 

For example, if we want to estimate the average height of **WNBA players**, we can use the **average height of players in a sample**. This approach allows us to make **inferences about the population** using available data.

In this notebook, we will explore two key theorems in classical statistics that justify this inference:
- **The Law of Large Numbers (LLN):** Ensures that as a sample size increases, the sample mean converges to the population mean.
- **The Central Limit Theorem (CLT):** States that the sampling distribution of the sample mean approaches a normal distribution, regardless of the population's original distribution.

## 1. Modeling a Sample

Consider a case where we observe a sample, $x = \{x_1, ..., x_N\}$, with $N$ observations.
We assume that the sample follows a **normal distribution**:

$$ x \sim \mathcal{N}(\mu, \sigma^2) $$

where:
- $\mu$ is the population mean,
- $\sigma^2$ is the variance of the normal distribution.

In econometrics and statistical inference, we typically **estimate the mean** of the sample, denoted as $\hat{\mu}$. However, due to random sampling, we expect:

$$ \hat{\mu} \neq \mu $$

Our objective is to understand how much our sample estimate deviates from the true mean.


## 2. Simulating a Sample in R

To understand sample variation, let's generate a small random sample from a normal distribution.

We assume:
- Mean ($\mu$) = **-3**
- Standard deviation ($\sigma$) = **5**
- Sample size ($N$) = **10**

```r
# Compute sample mean and standard deviation
sample_mean <- mean(x)
sample_sd <- sd(x)
```

## 3. The Distribution of Sample Means

We have seen that a single sample provides an estimate of the population mean, but this estimate varies depending on the sample. 

To understand this variability, let's consider the idea of **many imaginary samples**:
- If we take multiple samples from the same population, each will yield a **different** estimate of the mean.
- The **distribution of these sample means** tells us about the **uncertainty** in our estimate.
- As we increase the number of samples, the distribution of sample means will center around the true mean.

This leads to a fundamental concept in statistics: the **sampling distribution**.

## 4. Simulating Multiple Samples in R

Let's generate **500 different samples**, each of size $N = 10$, from a normal distribution with mean $-3$ and standard deviation $5$. 

We will then:
1. Compute the **mean** for each sample.
2. Plot the **distribution of sample means**.

```r
# Set parameters
set.seed(12345)
M <- 500  # Number of samples
N <- 10   # Sample size
```

```r
# Initialize matrix to store samples
sample_est <- matrix(NA, M, N)

# This creates an M x N matrix filled with NA values, where:
# - Rows (M) represent different samples
# - Columns (N) represent individual observations within each sample
```

```r
# Generate multiple samples
for (m in 1:M) {
  sample_est[m, ] <- rnorm(N, mean = -3, sd = 5)
}
```

```r
# Compute means of each sample
sample_means <- rowMeans(sample_est)
```

```r
# Plot histogram of sample means

# Set plot background to white
par(bg = "white")

hist(sample_means, breaks = 30, col = "lightblue",
     main = "Distribution of Sample Means",
     xlab = "Sample Mean")

# Add vertical red line for the average sample mean
abline(v = mean(sample_means), col = "red", lwd = 2.5)
```

```r
# Average of sample means
print(mean(sample_means))
```

# 5. The Law of Large Numbers (LLN)

The **Law of Large Numbers (LLN)** states that as the sample size $N$ increases, the sample mean **converges** to the true population mean:

$$ \lim_{N \to \infty} \frac{1}{N} \sum_{i=1}^{N} x_i = \mu $$

where:
- $x_i$ are independent observations.
- $\mu$ is the true mean of the population.

This theorem guarantees that, with **a sufficiently large sample size**, our sample mean $\hat{\mu}$ will be **very close** to the true mean $\mu$.

### **Example: Simulating the LLN**
We generate a very large sample ($N = 100,000$) from a **normal distribution** with:
- Mean $\mu = -3$
- Standard deviation $\sigma = 5$

We then compute the **sample mean** and compare it to the true value.

```r
# Set seed for reproducibility
set.seed(123456789)

# Define parameters
N <- 100000  # Large sample size
mu <- -3
sigma <- 5

# Generate random sample
x <- rnorm(N, mean = mu, sd = sigma)

# Compute sample mean
sample_mean <- mean(x)

# Print result
sample_mean  # Should be very close to -3
```

# 6. Central Limit Theorem (CLT) Simulation


The **Central Limit Theorem (CLT)** states that as the sample size increases, the **distribution of the sample mean** approaches a **normal distribution**, regardless of the original population distribution.

Formally, if $X_1, X_2, \dots, X_N$ are independent and identically distributed (i.i.d.) random variables with:
- Expected value: $E(X_i) = \mu$
- Variance: $Var(X_i) = \sigma^2$

Then, as $N \to \infty$,

$$ \sqrt{N} \left( \frac{1}{N} \sum_{i=1}^{N} X_i - \mu \right) \sim \mathcal{N}(0, \sigma^2) $$

This means:
- The **sample mean** converges to the **true mean**.
- The **sampling distribution of the sample mean** becomes normal for large $N$.
- The **variance of the sample mean decreases** as $N$ increases.



## Simulating the CLT with Different Sample Sizes
We will:
1. Draw **1000 random samples** from a **normal distribution** $\mathcal{N}(-3, 5^2)$.
2. Compute **sample means** for different sample sizes: $N = 10, 100, 1000, 10000$.
3. Plot the **density** of these sample means to observe the normality.

# 7. Approximation of the Limiting Distribution

The standard solution to not knowing the true values is to use an approximation based on the **analogy principle** (Manski, 1990).  
Thus, we simply replace the unknown **true mean** with its known analogy, the **sample mean**.  
Similarly, we replace the **unknown true variance** with the **sample variance**.



### Assumption: Approximate Distribution
We approximate the distribution of the sample mean using:

$$ \hat{\mu} \sim \mathcal{N} \left( \hat{\mu}, \frac{\hat{\sigma}^2}{N} \right) $$

where:

$$ \hat{\mu} = \frac{1}{N} \sum_{i=1}^{N} x_i $$

$$ \hat{\sigma}^2 = \frac{1}{N} \sum_{i=1}^{N} x_i^2 - \hat{\mu}^2 $$



### Interpretation
- The above assumption states that the **estimated mean** is distributed normally,  
  with a **mean equal to itself** and a **variance equal to the sample variance divided by \( N \)**.
- This approximation gives a **measure of uncertainty** that we can actually use.



### Challenges with This Approximation
There are **two problems** with this approach:

1. **Normality Assumption:**  
   - There is no strong reason to assume that the **sample mean follows a normal distribution**.  
   - The approximation **only holds for large \( N \)**, but not for small samples.

2. **Uncertainty in Variance Estimate:**  
   - The measure of uncertainty itself **depends on the sample** and may be **poorly estimated**.
   - The more **uncertainty there is**, the **less certain we are** about how much uncertainty actually exists.


# 7.1 Simulation of Approximate Distributions


Let's **simulate five different samples** and compute the corresponding **approximate distribution of the sample mean** from each sample.

We assume a normal distribution for the data and calculate:
- **Sample mean**: $\hat{\mu} = \frac{1}{N} \sum_{i=1}^{N} x_i$

- **Approximate standard deviation of the mean**:  
  $$
  \hat{\sigma}_{\hat{\mu}} = \frac{\hat{\sigma}}{\sqrt{N}}
  $$
  where $\hat{\sigma}$ is the sample standard deviation.


```r
# Set parameters
M <- 5   # Number of samples
N <- 10  # Sample size
```

```r
# Initialize matrix to store sample means and standard deviations
sample_means <- matrix(NA, M, 2)  

# Initialize empty vector for x-axis values
x2 <- NULL  
```

```r
# Loop to generate samples and compute mean & variance
for (m in 1:M) {
  x1 <- rnorm(N, mean = mu, sd = sigma)  # Generate random sample from N(mu, sigma^2)
  sample_means[m,] <- c(mean(x1), sd(x1) / sqrt(N))  # Store mean & std deviation of mean
  x2 <- c(x2, x1)  # Collect data points for sorting
}
```

# 8. Bootstrap

An alternative approach to representing **uncertainty** is called **bootstrapping**.  
This idea was developed by **Stanford statistician Brad Efron** and extends the **analogy principle** to its natural conclusion.

- If we need the **true distribution**, we use its **sample analog**, i.e., the **sample itself**.
- Using a computer, we can create **counterfactual pseudo-samples** by **re-drawing from the sample with replacement**.



## Example: Bootstrap Estimation
We illustrate bootstrapping using a small **sample of size \( N = 10 \)** from a normal distribution.

### R Code: Bootstrap Simulation

```r
# Set sample size
N <- 10
x1 <- rnorm(N, mean = mu, sd = sigma)  # Generate a sample from N(mu, sigma^2)

# Set number of bootstrap resamples
M <- 1000  

# Create matrix to store bootstrap sample means
bs_mean <- matrix(NA, M, 1)  

# Bootstrap loop: Resample with replacement
for (m in 1:M) {
  index_bs <- sample(1:N, size = N, replace = TRUE)  # Correct bootstrap resampling
  bs_mean[m,1] <- mean(x1[index_bs])  # Compute sample mean
}

# Compute density of bootstrap means
dens_bs <- density(bs_mean)

# Generate x-axis values for true normal density
x_vals <- seq(min(bs_mean), max(bs_mean), length.out = 1000)

# Compute true normal density using predefined mu and sigma
true_density <- dnorm(x_vals, mean = mu, sd = sigma / sqrt(N))

# Set plot background to white
par(bg = "white")

# Plot the bootstrap distribution (solid black line)
plot(dens_bs, col = "black", lwd = 3, type = "l",
     main = "Bootstrapped vs. True Normal Distribution",
     xlab = "Mean", ylab = "Density")

# Overlay the true normal distribution (dashed black line)
lines(x_vals, true_density * max(dens_bs$y) / max(true_density), 
      col = "black", lwd = 3, lty = 2)

# Add vertical line at true mean
abline(v = mu, col = "black", lwd = 3)

# Add legend for clarity
legend("topright", legend = c("Bootstrapped", "True Normal"), 
       col = c("black", "black"), lwd = 3, lty = c(1, 2))
```

# Intuition Behind Bootstrapping

## The Problem: Uncertainty in Estimation
When we calculate the **sample mean** from a small dataset, we know it’s **just one possible estimate**. If we took a different sample, we might get a **different mean**. The key question is:

> *How much does our estimate vary?*

Ideally, we would collect **many new samples from the population**, but that’s often **impossible** due to time, cost, or feasibility constraints.



## The Solution: Bootstrapping
Bootstrapping **simulates** what would happen if we took many new samples **without actually collecting more data**.

1. **Take a single dataset**: We start with **one observed sample** (e.g., $x_1, x_2, ..., x_N$).
2. **Resample with replacement**: We repeatedly create **new pseudo-samples** by drawing **randomly from our original sample**, allowing duplicates.
3. **Compute the statistic each time**: For each new pseudo-sample, we compute the **sample mean**.
4. **Build a distribution**: After many resamples, we get a **distribution of sample means**—this is our **bootstrap distribution**.


## What Does Bootstrapping Tell Us?
- **Estimate of Uncertainty**: The spread of the bootstrap distribution tells us **how much the sample mean varies**.
- **Confidence Intervals**: We can construct **95% confidence intervals** using the percentiles of the bootstrap distribution.
- **No Assumption of Normality**: Unlike traditional methods (which assume the sample mean follows a normal distribution), **bootstrapping works for any distribution**.



## Comparing to the True Normal Distribution
- **Theoretical Normal Distribution**: If the **true population distribution is normal**, then the **sample mean should be normally distributed** with:
  
  $$ \mathcal{N} \left( \mu, \frac{\sigma}{\sqrt{N}} \right) $$

- **Bootstrap Approximation**: Bootstrapping **estimates this distribution** **without assuming normality** by resampling from our given dataset.



## What the Graph Shows
- **Solid Line (Bootstrapped Distribution)**: This is the estimated distribution of the sample mean, based on resampling.
- **Dashed Line (True Normal Distribution)**: This is what theory predicts if the **original data is normally distributed**.

### Key Observations
1. **If bootstrapping is accurate**, the two curves should be **similar**.
2. **With small samples**, the bootstrap distribution might be **wider or slightly skewed**.
3. **As sample size increases**, bootstrapping gets **closer to the true normal distribution**.



## Final Takeaway
Bootstrapping is like **re-running an experiment thousands of times** using the **data we already have**. It helps us **understand the variability** of our estimates **without assuming normality or collecting new data**.
